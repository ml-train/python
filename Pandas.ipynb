```python
plt.figure(figsize=(12,4))
plt.plot(['Jan','Feb','Mar'],[12,34,19.6],label ='2022',color='teal',marker='o',lw=3)
plt.plot(['Jan','Feb','Mar'],[11,27,23],label='2021',linestyle='dashed',marker='o')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Sales by Month")
plt.legend(loc="best")
plt.show()
```

```python
df['Vehicle type'].unique()
```

```python
report=pd.pivot_table(data=df,index='Manufacturer',columns='Vehicle type',values='Sales in thousands',aggfunc='sum').fillna(0)
```

```python
plt.figure(figsize=(14,8))
plt.bar(report.index,report.Car)
plt.bar(report.index,report.Passenger,bottom=report.Car)
plt.xlabel("Manufacturer")
plt.ylabel("Sales")
plt.title("Sales by Manufacturer")
plt.xticks(rotation=45)
plt.show()
```

```python
rp=report[report.Car>0].reset_index()
```

```python
plt.figure(figsize=(14,8))
plt.bar(rp.index,rp.Car,width=0.5)
plt.bar(rp.index+0.5,rp.Passenger,width=0.5)
plt.xlabel("Manufacturer")
plt.ylabel("Sales")
plt.title("Sales by Manufacturer")
plt.xticks(rotation=45)
plt.show()
```
